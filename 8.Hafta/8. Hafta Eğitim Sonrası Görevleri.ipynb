# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# 1.Problemin Çözümünü Buraya Yazınız
girişSayısı = int(input("giriş sayısı giriniz :  "))

def vki_hesapla (isim, yas, uzunluk, ağırlık):
    Uzunluk= uzunluk / 100
    vki = ağırlık / (Uzunluk ** 2)
    if vki < 18.5:
        vki_kategori = "zayıf"
    elif vki < 25:
        vki_kategori = "Normal kilolu"
    elif vki < 30:
        vki_kategori = "fazla kilolu"
    elif vki < 40:
        vki_kategori = "Obez"
    else:
        vki_kategori = "ileri derece obez"
    return {"isim": isim, "yas": yas, "uzunluk": uzunluk, "ağırlık": ağırlık, "vki": vki, "vki_kategori": vki_kategori}

for i in range(girişSayısı):
    isim = input("isim giriniz : ")
    yas = int(input("yaş giriniz : "))
    uzunluk = float(input("uzunluk giriniz (cm) : "))
    ağırlık = float(input("ağırlık giriniz (kg) : "))
    veri = vki_hesapla(isim, yas, uzunluk, ağırlık)
    vki_kategori = veri["vki_kategori"]
    with open(f"{vki_kategori}.txt", "a") as f:
        f.write(str(veri) + "\n")


```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
# 2.Problemin Çözümünü Buraya Yazınız
import datetime

def menuyu_goruntule():
    print("YetBar Sağlıklı İçecek Otomatı")
    print("******************************")
    print("1. Menüyü görüntüle")
    print("2. Satın alma")
    print("3. Çıkış")

def urunleri_goruntule(urunler):
    print("Menü:")
    for urun, fiyat in urunler.items():
        print("- {} - {} TL".format(urun, fiyat))

def urun_fiyati_hesapla(urunler, secilen_urunler):
    toplam_fiyat = 0
    for urun in secilen_urunler:
        if urun in urunler:
            toplam_fiyat += urunler[urun]
    return toplam_fiyat

def siparis_kaydet(tarih, fiyat):
    with open("siparisler.txt", "a") as dosya:
        dosya.write("{} - {} TL\n".format(tarih, fiyat))

urunler = {
    "Yeşil Smoothie": 8,
    "Meyve Smoothie": 10,
    "Protein Shake": 12,
    "Detoks Suyu": 6,
    "Süper Enerji İçeceği": 14
}

while True:
    menuyu_goruntule()
    secim = input("Yapmak istediğiniz işlemi seçin: ")

    if secim == "1":
        urunleri_goruntule(urunler)
    elif secim == "2":
        secilen_urunler = []
        while True:
            urun = input("Satın almak istediğiniz ürünü girin (Çıkış için 'q' tuşuna basın): ")
            if urun == "q":
                break
            secilen_urunler.append(urun)
        toplam_fiyat = urun_fiyati_hesapla(urunler, secilen_urunler)
        tarih = datetime.datetime.now().strftime("%d/%m/%Y %H:%M:%S")
        siparis_kaydet(tarih, toplam_fiyat)
        print("Toplam fiyat: {} TL".format(toplam_fiyat))
        print("Satın alma işlemi tamamlandı.")
    elif secim == "3":
        print("Çıkış yapılıyor...")
        break
    else:
        print("Geçersiz bir seçim yaptınız. Lütfen tekrar deneyin.")
3
```
