# Problem 1 Aradaki Tam Sayılar

Kullanıcıdan iki adet sayı alınız, bu sayılar arasındaki tam sayıları ekrana yazdıran bir program yazınız. 

```python
# 1.Problemin Çözümünü Buraya Yazınız
ilk_sayi = int(input("İlk sayıyı girin: "))
ikinci_sayi = int(input("İkinci sayıyı girin: "))

# İlk sayıdan ikinci sayıya kadar olan tam sayıları yazdırma
for sayi in range(ilk_sayi, ikinci_sayi + 1):
    print(sayi)
```

# Problem 2 Faktöriyel

Kullanıcıdan bir sayı alınız ve alınan sayının faktöriyelini hesaplayan bir program yazınız. 


```python
# 2.Problemin Çözümünü Buraya Yazınız
sayi = int(input("Bir sayı girin: "))

faktoriyel = 1
for i in range(1, sayi + 1):
    faktoriyel *= i

print("Faktöriyel:", faktoriyel)
```

# Problem 3 Virgül ile Ayırma

Kullanıcıdan bir metin halinde virgül ile ayrılmış sayılar alınız ve aldığınız her bir sayıyı listeye atayınız, daha sonra listedeki her elemanı toplayıp ekrana bastıran bir program yazınız.

<h4>Sample Input</h4>
<p>
Lutfen sayilari virgul ile ayrirarak giriniz: 5,14,7,2,1
</p>
<h4>Output</h4>

Toplam: 29

```python
# 3.Problemin Çözümünü Buraya Yazınız
sayilar = input("Virgülle ayrılmış sayıları girin: ")
sayi_listesi = sayilar.split(",")

toplam = 0
for sayi in sayi_listesi:
    toplam += int(sayi)

print("Toplam:", toplam)
```

# Problem 4 Büyük Sayı

Kullanıcıdan alınan iki sayıdan büyük olanı bulup, ekrana yazdıran bir fonksiyon yazınız.

```python
# 4.Problemin Çözümünü Buraya Yazınız
def buyuk_sayiyi_bul(sayi1, sayi2):
    if sayi1 > sayi2:
        print("Büyük sayı:", sayi1)
    elif sayi2 > sayi1:
        print("Büyük sayı:", sayi2)
    else:
        print("Sayılar birbirine eşit.")

# Kullanıcıdan sayıları alıp fonksiyonu çağırma
sayi1 = int(input("Birinci sayıyı girin: "))
sayi2 = int(input("İkinci sayıyı girin: "))
buyuk_sayiyi_bul(sayi1, sayi2)
```

# Problem 5 Ürün Kayıt Uygulaması
Programımız ilk olarak kullanıcıdan, kaydı yapılacak ürünlerin sayısını isteyecek. Ardından yazılmış olan ürün sayısı kadar ürün ismi ve ürün fiyatı bilgisini kullanıcıdan alacak. Son olarak ise ürünleri ekrana yazdıracağız.

<h4>Sample Input:</h4>
<p>
Lutfen kac adet urun gireceginizi yaziniz: 3
<br/>
1. Urununuzun ismini giriniz: Mentos
<br/>
1. Urununuzun fiyatini giriniz: 1
<br/>
2. Urununuzun ismini giriniz: Meybuz
<br/>
2. Urununuzun fiyatini giriniz: 0.25
<br/>
3. Urununuzun ismini giriniz: Uzun Şeker
<br/>
3. Urununuzun fiyatini giriniz: 0.75
<br/>
<h4>Output:</h4>
1. Urun: Mentos - 1 TL
<br/>
2. Urun: Meybuz - 0.25 TL
<br/>
3. Urun: Uzun Şeker - 0.75 TL
</p>

```python
# 5.Problemin Çözümünü Buraya Yazınız

adet = int(input("Lütfen kaç adet ürün gireceğinizi yazın: "))

urunler = []
for i in range(1, adet + 1):
    isim = input(f"{i}. Ürününüzün ismini giriniz: ")
    fiyat = float(input(f"{i}. Ürününüzün fiyatını giriniz: "))
    urunler.append((isim, fiyat))

print("Ürünler:")
for i, urun in enumerate(urunler, 1):
    isim, fiyat = urun
    print(f"{i}. Ürün: {isim} - {fiyat} TL")
```

# Problem 6 Türkçe Karakter Çevirici
Programa girilen dosya ismindeki Türkçe karakterleri, İngilizce'ye çeviren programı yazınız.

<h4>Sample Input:</h4>
<p>
Lutfen dosya ismi giriniz: fotoğraf.jpeg
<br>
Dosya ismi: fotograf.jpeg
<br>
<br>
Lutfen dosya ismi giriniz: ışık.png
<br>
Dosya ismi: isik.png
</p>

```python
# 6.Problemin Çözümünü Buraya Yazınız
def convert_to_english(filename):
    turkish_chars = "çÇğĞıİöÖşŞüÜ"
    english_chars = "cCgGiIoOsSuU"
    converted_name = ""

    for char in filename:
        if char in turkish_chars:
            index = turkish_chars.index(char)
            converted_name += english_chars[index]
        else:
            converted_name += char

    return converted_name


dosya_ismi = input("Lütfen dosya ismini giriniz: ")
converted_name = convert_to_english(dosya_ismi)
print("Dosya ismi:", converted_name)
```

# Problem 7 Sayı Tahmin Oyunu ama Bilgisayar Tahmin Ediyor :)
Sayı tahmin oyunu: İki oyuncudan birinin, 0 ile 100 arasında bir sayı tutması ile başlar. Diğer oyuncu ise tahminler yaparak tutulan sayıyı bulmaya çalışır ve yaptığı her tahminde sayıyı tutmuş olan oyuncudan YUKARI veya AŞAĞI olarak geri bildirim alır. Sayıyı bildiğinde ise rakip oyuncu kaç tahminde (kaç denemede) bildiğini söyler.

Senden yazmanı beklediğimiz program ise Python'ın tahminleri yapan tarafta olacağı bir oyun oluşturman. Yani sen 0 ile 100 arasında bir sayı tutacaksın ardından program senin tuttuğunuz sayıyı bulmaya çalışacak.
<h4>Sample Input:</h4>
Bu örnekte aklınızdan 90 tuttuğunuzu varsayalım.
<p>
Oyun başlıyor lutfen aklindan bir sayi tut! Tuttun mu, hazır mısın? : 1
<br>
Tahmin: 50
<br>
Durum: Yukari
<br>
Tahmin: 75
<br>
Durum: Yukari
<br>
Tahmin: 87
<br>
Durum: Yukari
<br>
Tahmin: 93
<br>
Durum: Asagi
<br>
Tahmin: 90
<br>
Durum: Bildin
</p>

```python
# 7.Problemin Çözümünü Buraya Yazınız
import random

def tahmin_et(baslangic, bitis):
    tahmin_sayisi = 0
    durum = None
    
    while durum != "Bildin":
        tahmin = random.randint(baslangic, bitis)
        print("Tahmin:", tahmin)
        durum = input("Durum (Yukari/Aşağı/Bildin): ")
        
        tahmin_sayisi += 1
        
        if durum == "Yukarı":
            baslangic = tahmin + 1
        elif durum == "Aşağı":
            bitis = tahmin - 1
    
    print("Tebrikler! Sayıyı {} denemede buldum.".format(tahmin_sayisi))

oyun_baslangic = 0
oyun_bitis = 100

print("Oyun başlıyor! Bir sayı tutun ve hazır olduğunuzda başlayalım.")
input("Tuttunuz mu, hazır mısınız? (Enter tuşuna basın)")

tahmin_et(oyun_baslangic, oyun_bitis)
```

# Problem 8 Net Yaş Hesaplayıcı

Kullanıcının bugünün güncel tarihine kadar yaşadığı ay, gün, yıl süresini hesaplayan bir Python kodu yazınız. (Her ay 30 gün kabul edilmeyecektir. Kodu yazarken hangi ayın kaç gün olduğuna [Şubat ayının 29 gün olduğu yıllar dahil] dikkat edin.)

```python
# 8.Problemin Çözümünü Buraya Yazınız
from datetime import date

bugun = date.today()

gun = int(input("Doğduğunuz günü girin: "))
ay = int(input("Doğduğunuz ayı girin: "))
yil = int(input("Doğduğunuz yılı girin: "))

dogum_tarihi = date(yil, ay, gun)

yas = bugun - dogum_tarihi

yas_yil = yas.days // 365
yas_ay = (yas.days % 365) // 30
yas_gun = (yas.days % 365) % 30

print("Yaşadığınız süre: {} yıl, {} ay, {} gün".format(yas_yil, yas_ay, yas_gun))
```

# Problem 9 Tek-Çift

Kullanıcıdan 'q' tuşuna basana kadar rastgele sayılar alın. Tek olanları bir listeye, çift olanları bir başka listeye büyükten küçüğe sıralanacak şekilde atın. Listeleri ekrana yazdıran bir kod yazın.

```python
# 9.Problemin Çözümünü Buraya Yazınız
tek_sayilar = []
cift_sayilar = []

while True:
    sayi = input("Bir sayı girin ('q' tuşuna basarak çıkış yapın): ")
    
    if sayi == 'q':
        break
    
    sayi = int(sayi)
    if sayi % 2 == 0:
        cift_sayilar.append(sayi)
    else:
        tek_sayilar.append(sayi)

tek_sayilar.sort(reverse=True)
cift_sayilar.sort(reverse=True)

print("Tek Sayılar:", tek_sayilar)
print("Çift Sayılar:", cift_sayilar)
```

 # Problem 10 Altıgen Çizimi

Kullanıcıdan altıgenin kaç satırdan oluşacağı ve orta satırında kaç karakter olacağı bilgisini alarak '*' karakteri ile bir altıgen çizin. (For loop kullanabilirsiniz.)

```python
# 10.Problemin Çözümünü Buraya Yazınız
satir_sayisi = int(input("Altıgenin kaç satırdan oluşacağını girin: "))
orta_satir_uzunlugu = int(input("Orta satırın uzunluğunu girin: "))

# Yukarıdaki kısım
for i in range(1, satir_sayisi + 1):
    bosluk_sayisi = satir_sayisi - i
    print(" " * bosluk_sayisi, end="")
    print("* " * i)

# Orta satır
print("* " * orta_satir_uzunlugu)

# Aşağıdaki kısım
for i in range(satir_sayisi, 0, -1):
    bosluk_sayisi = satir_sayisi - i
    print(" " * bosluk_sayisi, end="")
    print("* " * i)
```
