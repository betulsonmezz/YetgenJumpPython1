# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
# 1.Problemin Çözümünü Buraya Yazınız
class Vehicle:
    def __init__(self, yapılış, model, yil):
        self.yapılış = yapılış
        self.model = model
        self.yil = yil

class Bus(Vehicle):
    def __init__(self, yapılış, model, yil):
        super().__init__(yapılış, model, yil)
        self.oturmaKapasitesi = 50

class Bus(Vehicle):
    def oturmaKapasitesi(self, kapasite=50):
        self.kapasite = kapasite
        print(f" yolcu kapasitesi: {self.kapasite}")


my_bus = Bus("Toyota", "Coaster", 2022)
my_bus.oturmaKapasitesi()  
````
